```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os

def ORB():

    root = os.getcwd()
    imgPath = "/content/4.PNG"
    #imgPath = os.path.join(root, "Images/4.PNG")
    imgGray = cv.imread(imgPath, cv.IMREAD_GRAYSCALE)

    if imgGray is None:
        print(f"Error: Could not load image at path: {imgPath}")
        print("Please ensure you have uploaded the image and the path is correct.")
        return

    orb = cv.ORB_create()
    key_points = orb.detect(imgGray, None)
    key_points, _ = orb.compute(imgGray, key_points)
    print(f"Detected {len(key_points)} key points.")


    imgGray = cv.drawKeypoints(imgGray, key_points, imgGray,
        flags=cv.DrawMatchesFlags_DRAW_RICH_KEYPOINTS
    )

    plt.figure()
    plt.imshow(imgGray)
    plt.show()

if __name__ == '__main__':
  ORB()
```
